```python
bins = [0, 100000, 300000, float("inf")]
labels = ["Low", "Mid", "High"]
df["Income Band"] = pd.cut(df["Estimated Income"],bins = bins, labels = labels, right = False)

df["Income Band"].value_counts().plot(kind="bar")
```

```python
#Examine the date catogery wise

category = df[['BRId', 'GenderId', 'IAId', 'Fee Structure','Loyalty Classification','Amount of Credit Cards', 'Occupation', 'Income Band', 'Properties Owned','Risk Weighting', 'Age', 'Nationality',]]
for col in category:
    print(f"values counts for {col}:")
    display(df[col].value_counts())
```

```python
for i, prep in enumerate(df[['BRId', 'GenderId', 'IAId', 'Fee Structure','Loyalty Classification','Amount of Credit Cards','Occupation', 'Income Band', 'Properties Owned','Risk Weighting', 'Age', 'Nationality' ]]):
    mt.figure(i)
    sns.countplot(data=df, x=prep, hue= "Nationality")
```

```python
for col in category:
    if col == "Occupation":
        continue
    mt.figure(figsize=(5,2))
    sns.histplot(df[col])
    mt.title("Histogram of the Occupation")
```

```python
numercial_col = ['Estimated Income', 'Superannuation Savings','Credit Card Balance', 'Bank Loans','Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account', 'Business Lending', 'Income Band']
mt.figure(figsize= (20,20))
for i, col in enumerate(numercial_col):
    mt.subplot(4, 3, i+1)
    sns.histplot(df[col],kde=True)
    mt.title(col)
mt.show()
```

```python
numercial_col = ['Estimated Income', 'Superannuation Savings','Credit Card Balance', 'Bank Loans','Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numercial_col].corr()

mt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True, cmap="coolwarm",fmt=".2f") 
mt.title("Heatmap")
mt.show()
```

```python
df["Income band num"] = df["Income Band"].map({"Low":1, "Mid":2, "High":3})
numercial_col = ['Estimated Income', 'Superannuation Savings','Credit Card Balance', 'Bank Loans','Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account', 'Business Lending', 'Income band num']

correlation_matrix = df[numercial_col].corr()

mt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix, annot=True, cmap='crest', fmt=".2f")
mt.title("Correlation Matrix")
mt.show()
```
